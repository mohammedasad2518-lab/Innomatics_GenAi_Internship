**_1: Mobile Recharge Validation System_**

```python
# Function to validate recharge amount
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False, "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return False, "Invalid plan. Available plans are 199, 299, 399, 599."
    else:
        return True, "Recharge Successful!"


# Retry mechanism using while loop
while True:
    amount = int(input("Enter recharge amount: ₹"))

    is_valid, message = validate_recharge(amount)
    print(message)

    if is_valid:
        break
    else:
        print("Please try again.\n")
```

**_ 2: Inventory Reorder Alert System_**

```python
# Function to check inventory stock levels
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Inventory data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 30
}

# Check inventory status
check_inventory(inventory)
```

**_3: Student Result Processing System_**

```python
# Function to process student result
def process_result(marks):
    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    # Determine result
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Student marks
student_marks = [45, 60, 70, 55]

# Get and display result
result = process_result(student_marks)
print("Result:", result)
```

**_4: Cab Fare Estimator with Retry Option_**

```python
# Function to calculate cab fare
def calculate_fare(distance_km, is_peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance_km * per_km_rate)

    if is_peak:
        fare += fare * 0.25

    return fare


# Retry mechanism using while loop
while True:
    distance = float(input("Enter distance in km: "))

    peak_input = input("Is it peak hour? (yes/no): ").lower()
    is_peak = peak_input == "yes"

    total_fare = calculate_fare(distance, is_peak)
    print(f"Total Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

**_5: Employee Attendance Eligibility Checker _**

```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_days = 0

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / len(attendance)) * 100

    # Determine eligibility
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Attendance data
employee_attendance = ["P", "P", "A", "P", "P", "A", "P"]

# Check and display result
result = check_attendance(employee_attendance)
print("Attendance Status:", result)
```

**_6: Password Strength Checker_**

```python
# Function to check password strength
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_characters = "@#$"

    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password"

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example usage
password = "Pass@123"
result = check_password_strength(password)
print("Password Strength:", result)
```
